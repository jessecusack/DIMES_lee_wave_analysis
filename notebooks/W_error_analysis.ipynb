$W_w$ error analysis
====================

Having estimated the vertical water velocity $W_w$ using an optimised float buoyancy model I now attempt to quantify the error on the estimates.

Approaches
----------

James: "There will likely be a peak around $N$ (which will be broad because of the range of stratification values in the profile but probably is real internal wave signal). If you see a white spectrum of level $P$ at higher frequencies up to the Nyquist ($f_{Ny}$), the RMS $W_w$ noise estimate is $\sqrt{P f_{Ny}}$."

Alberto: "... look at vertical wavenumber spectra of $W_w$ and compare them with the Garrett-Munk model of the internal wave field. We should see that above some characteristic high wavenumber noise dominates our signal, and estimate the $W_w$ magnitude typically associated with wavenumbers  higher than that from the integrated $W_w$ variance in that part of the spectrum."

These approaches sound equivalent to me.

Setup
-----

The data for floats 4976 and 4977 is loaded along with some useful modules for signal processing. The vertical water velocity is calculated using previously optimised model parameters and the adiabatically levelled buoyancy (using 300 dbar bin width) is also loaded.

```python
import scipy.signal as sig
import scipy.optimize as op
import matplotlib.pyplot as plt
import pylab as pyl
import numpy as np
import emapex
import utils

%qtconsole
%matplotlib inline
    
reload(emapex)

E76 = emapex.EMApexFloat('../../data/EM-APEX/allprofs11.mat', 4976)
E76.apply_w_model('../../data/EM-APEX/4976_fix_p0k0M_fit_info.p')
E76.generate_regular_grids()

E77 = emapex.EMApexFloat('../../data/EM-APEX/allprofs11.mat', 4977)
E77.apply_w_model('../../data/EM-APEX/4977_fix_p0k0M_fit_info.p')
E77.generate_regular_grids()
```

First look at how well the cost function holds up with depth and across smaller groups of profiles. The code here is binning all the $W_w$ data into 50m depth bins and also 3 day bins then computing some statistics. 

```python
# Look over all profiles.
hpids = np.arange(1,1000)

for Float in [E76, E77]:
    
    __, Ww = Float.get_timeseries(hpids, 'r_Ww')  # Water velocity from model
    __, z = Float.get_timeseries(hpids, 'r_z')  # Depth
    __, UTC = Float.get_timeseries(hpids, 'r_UTC')  # Time
    
    # 3 day bins.
    tbins = np.arange(UTC[0], UTC[-1], 3)
    # 50m depth bins
    zbins = np.arange(-1500, 0, 50)
    
    # bin data
    tidxs = np.digitize(UTC, tbins)
    zidxs = np.digitize(z, zbins)
    
    tN = len(tbins)
    zN = len(zbins)
    
    ## Calculate stats.
    # 3 day.
    UTC_mean = np.array([np.nanmean(UTC[tidxs == i]) for i in xrange(1, tN+1)])
    cost = np.array([np.nansum(Ww[tidxs == i]**2) for i in xrange(1, tN+1)])
    Ww_mean = np.array([np.nanmean(Ww[tidxs == i]) for i in xrange(1, tN+1)])
    Ww_std = np.array([np.nanstd(Ww[tidxs == i]) for i in xrange(1, tN+1)])
    
    time = utils.datenum_to_datetime(UTC_mean)
    
    plt.figure()
    plt.semilogy(time, cost, time, np.abs(Ww_mean), time, Ww_std)
    plt.title("Float {}".format(Float.floatID))
    plt.xlabel('Time')
    plt.xticks(rotation=45)
    plt.legend(['Cost (m$^2$ s$^{-2}$)', 'Mean $|W_w|$ (m s$^{-1}$)', 'Std $W_w$ (m s$^{-1}$)'], loc=0)
    
    print("Float {}, std of mean value with time: {} mm s-1".format(Float.floatID, 1000.*np.nanstd(Ww_mean)))
    
    # 50 m.
    z_mean = np.array([np.nanmean(z[zidxs == i]) for i in xrange(1, zN+1)])
    cost = np.array([np.nansum(Ww[zidxs == i]**2) for i in xrange(1, zN+1)])
    Ww_mean = np.array([np.nanmean(Ww[zidxs == i]) for i in xrange(1, zN+1)])
    Ww_std = np.array([np.nanstd(Ww[zidxs == i]) for i in xrange(1, zN+1)])
       
    plt.figure()
    plt.semilogx(cost, z_mean, np.abs(Ww_mean), z_mean, Ww_std, z_mean)
    plt.title("Float {}".format(Float.floatID))
    plt.ylabel('Depth (m)')
    plt.legend(['Cost (m$^2$ s$^{-2}$)', 'Mean $|W_w|$ (m s$^{-1}$)', 'Std $W_w$ (m s$^{-1}$)'], loc=0)
    
    print("Float {}, std of mean value with depth: {} mm s-1".format(Float.floatID, 1000.*np.nanstd(Ww_mean)))
      
    
    
```

The standard deviation of the mean with depth is about 2 mm s$^{-1}$, and with time is about 5 mm s$^{-1}$

Now lets look at the vertical velocity power spectrum $E(m)$ where $m$ is vertical wavenumber. Given a depth grid of 5 m the Nyquist wavenumber is equal to 10 m. It is assumed that noise is introduced by a random error in the pressure (depth) measurement. The power spectra of 50 profiles are averaged.

```python
scaling='density'

# Generate float vertical velocity with random depth error.
# Fake profile
dze = 5.
dt = dze/0.11  # 0.11 is a tpyical descent/ascent speed.
zvals = np.arange(-1400., -150, dze)
N = 50
Pe = np.empty((zvals.size/2, N))

# Use an arbitrary profile for the sizes.
pfl = E76.get_profiles(155)
surface = pfl.r_z > -150.
z = pfl.r_z[~surface]
Pw76 = np.empty((z.size/2+1, N))
Pw77 = np.empty((z.size/2+1, N))

for i in xrange(N):
    ze = zvals + 0.13*np.random.rand(zvals.size)
    We = np.diff(ze)/dt
    me, Pe[:, i] = sig.periodogram(We, 1./dze, scaling=scaling)
    
    # Pick a random profile.
    pfl76 = E76.get_profiles(80+i)
    pfl77 = E77.get_profiles(70+i)

    # Chop out the top 150 m.
    Ww76 = pfl76.r_Ww[~surface]
    Ww77 = pfl77.r_Ww[~surface]

    m, Pw76[:,i] = sig.periodogram(Ww76, 1./dze, scaling=scaling)
    __, Pw77[:,i] = sig.periodogram(Ww77, 1./dze, scaling=scaling)
    
mean_Pe = np.mean(Pe, axis=1)
mean_Pw76 = np.mean(Pw76, axis=1)
mean_Pw77 = np.mean(Pw77, axis=1)

# Vertical velocity as been regridded to every 5 m.
dz = 5.
dm = 1./dz

# Remove the 0 wavenumber (infinite wavelength)
m, mean_Pw76, mean_Pw77, me, mean_Pe = m[1:], mean_Pw76[1:], mean_Pw77[1:], me[1:], mean_Pe[1:]
plt.figure()
plt.loglog(m, mean_Pw76)
plt.loglog(m, mean_Pw77)
plt.loglog(me, mean_Pe)
plt.ylabel('Mean power spectral density (m$^3$ s$^{-2}$)')
plt.xlabel('Wavenumber (m$^{-1}$)')
plt.legend(['4976', '4977', 'noise'], loc=0)
plt.grid()

mmin = np.mean([m[mean_Pw76.argmin()], m[mean_Pw77.argmin()]])
pmin = np.mean([mean_Pw76.min(), mean_Pw77.min()])
print('Wavenumber at which noise begins to dominate: {} m-1'.format(mmin))
print('Wavelength: {} m'.format(1./mmin))
print('Power at minima: {:1.1e} m3 s-2'.format(pmin))
print('Error estimate: {} m s-1'.format(np.sqrt(pmin*m.max())))

# Try fitting power law to wavenumbers < 5e-2 m-1
#idxs = m < 5e-2
#pfit = np.polyfit(np.log(m[idxs]), np.log(Pw[idxs]), 1)
#plt.loglog(m, m**pfit[0]/np.mean(m[idxs]**pfit[0]/Pw[idxs]))
```

Applying James's suggestion to a noisy power density at about $5 \times 10^{-5}$ m$^3$ s$^{-2}$ gives a noise error of order 7 mm s$^{-1}$. Which is about the same suggested by standard deviation in the mean $W_w$ with time. The noise spectra was aligned with the noisy part of the velocity spectra by trail and error and it was found that an error in the depth measurement of about 15 cm (or equivalently the pressure measurement error of about 0.15 dbar) is sufficient to generate a similar power at high wavenumbers. 

This analysis suggests that the error on $W_w$ is about 7 mm s$^{-1}$ not including errors in the model used to deduce it. It also suggests that noise begins to dominate on scales of 10 m.
