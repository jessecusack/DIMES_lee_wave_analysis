Sections of temperature, salinity and potential density
-------------------------------------------------------

```python
hpids = np.arange(1, 600)

vars = ['T', 'S', 'rho_1']
texvars = ['$T$', '$S$', r'$\sigma_1$']
units = [r'$^\circ$C', 'PSU', 'kg m$^{-3}$']
clims = [(2., 6.), (34., 34.6), (31.2, 32.3)]
cmaps = [plt.get_cmap(s) for s in ['rainbow', 'summer', 'spring']]

for Float in [E76, E77]:

    __, idxs = Float.get_profiles(hpids, ret_idxs=True)
    z = getattr(Float, 'z')[:, idxs].flatten(order='F')
    d = getattr(Float, 'dist_ctd')[:, idxs].flatten(order='F')

    Xg, Zg, rho_1g = Float.get_griddata_grid(hpids, 'dist_ctd', 'z', 'rho_1')
    rho_1g -= 1000.
    
    for var, texvar, unit, clim, cmap in zip(vars, texvars, units, clims,
                                             cmaps):

        C = getattr(Float, var)[:, idxs].flatten(order='F')

        if var == 'rho_1':
            C -= 1000.

        nans = np.isnan(z) | np.isnan(d) | np.isnan(C)

        plt.figure(figsize=(12, 4))
        plt.scatter(d[~nans], z[~nans], c=C[~nans], edgecolor='none',
                    cmap=cmap)
        cbar = plt.colorbar(orientation='horizontal', extend='both')
        cbar.set_label(texvar+' ('+unit+')')
        plt.clim(*clim)

        plt.contour(*Float.get_griddata_grid(hpids, 'dist_ctd', 'z', var),
                    N=6, colors='k')
        plt.contour(Xg, Zg, rho_1g, N=6, colors='w')
        
        plt.ylim(np.min(z[~nans]), np.max(z[~nans]))
        plt.xlim(np.min(d[~nans]), np.max(d[~nans]))
        plt.xlabel('Distance (km)')
        plt.ylabel('Depth (m)')
        title_str = ("Float {}").format(Float.floatID)
        plt.title(title_str)
```

Lets zoom in on those and also look at the bathymetry which is interpolated from Smith and Sandwell. 

```python
import sandwell

hpids = np.arange(10, 50)
bathy_file = '../../data/sandwell_bathymetry/topo_17.1.img'
vars = ['Ww', 'U_abs', 'V_abs']
zvars = ['z', 'zef', 'zef']
dvars = ['dist_ctd', 'dist_ef', 'dist_ef']
texvars = ['$W_w$', '$U$', '$V$']
clims = [(-10., 10.), (-100., 100.), (-100, 100.)]

for Float in [E76, E77]:

    __, idxs = Float.get_profiles(hpids, ret_idxs=True)

    for var, zvar, dvar, texvar, clim in zip(vars, zvars, dvars, texvars,
                                             clims):

        V = getattr(Float, var)[:, idxs].flatten(order='F')
        z = getattr(Float, zvar)[:, idxs].flatten(order='F')
        d = getattr(Float, dvar)[:, idxs].flatten(order='F')

        plt.figure(figsize=(5, 7))
        plt.scatter(d, z, s=50, c=V*100., edgecolor='none', cmap=bwr)
        cbar = plt.colorbar(orientation='horizontal', extend='both')
        cbar.set_label(texvar+' (cm s$^{-1}$)')
        plt.clim(*clim)

        plt.xlim(np.nanmin(d), np.nanmax(d))
        plt.xlabel('Distance (km)')
        plt.ylabel('Depth (m)')
        title_str = ("Float {}").format(Float.floatID)
        plt.title(title_str)

        lons = Float.lon_start[idxs]
        lats = Float.lat_start[idxs]
        dist = Float.dist[idxs]
        bathy = sandwell.interp_track(lons, lats, bathy_file)
        plt.plot(dist, bathy, 'k', linewidth=3)

        plt.ylim(np.nanmin(bathy), np.nanmax(z))

    pf.track_on_bathy(Float, hpids, bathy_file=bathy_file)
```

To get the zig-zag scatter in the above plots I have assumed that the float moves at a constant horizontal speed (which is determined from GPS and time interpolation) as it descends/ascends. I think this is probably more realistic that assuming that it goes up and down vertically. If the distance that a float travels from the start of an ascent/descent to the end is $X$ and the time it takes is $T$ then its average speed is $X/T$. I add this speed (after decomposing into zonal and meridional components) to get a first order estimate of the absolute horizontal water velocity from the EM sensors, which is plotted below:

The floats are in a region of strong eastward flow of order 1 m s$^{-1}$ and slow meridional flow. Given that they are located toward the north of the Drake Passage at this time they appear to be carried by a strong Subantarctic/Polar frontal jet. The colocation of a sea mount (width of order 20 km) and strong zonal flow raises the strong suspicion that the large vertical motions are some kind of mountain wave.

Power spectrum of full time series
----------------------------------

```python
t, w = E76.get_timeseries(np.arange(1, 600), 'Ww')
t *= 60*60*24

plt.figure(figsize=(16,4))
plt.plot(t, w)

fs = 1./np.mean(np.diff(t))
freq, P = sig.welch(w, fs=fs, nperseg=2048)

plt.figure()
plt.loglog(freq, P)
# Where I think N is...
plt.vlines(2.8e-3*2*np.pi, *plt.ylim())
# Another peak I don't know... I guess it could also be N.
# Perhaps the float moved into another regime across a front and the stratification changed?
plt.vlines(1.9e-3*2*np.pi, *plt.ylim())
# Where I think f is...
plt.vlines(gsw.f(56)*2*np.pi, *plt.ylim())
plt.grid()
```

```python
t, U = E76.get_timeseries(np.arange(1, 600), 'U_abs')
t *= 60*60*24

nans = np.isnan(U)

plt.figure(figsize=(16,4))
plt.plot(t, w)

fs = 1./np.mean(np.diff(t))
freq, P = sig.welch(U[~nans], fs=fs, nperseg=1024)

plt.figure()
plt.loglog(freq, P)
# Where I think N is...
plt.vlines(2.8e-3*2*np.pi, *plt.ylim())
# Another peak I don't know... I guess it could also be N.
# Perhaps the float moved into another regime across a front and the stratification changed?
plt.vlines(1.9e-3*2*np.pi, *plt.ylim())
# Where I think f is...
plt.vlines(gsw.f(56)*2*np.pi, *plt.ylim())
plt.grid()
```
