I am following the methodology presented by Vallis in the new chapters of his GFD book and Durran too. It is as follows:

Consider a topographic obstacle $h(x)$ over a region can be represented as a Fourier series

\begin{equation}
h(x) = \sum_k h_k e^{ikx}
\end{equation}

where $h_k$ are the complex Fourier coefficients found by taking the fourier transform of $h(x)$. Linear theory for steady two-dimensional Boussinesq ﬂow over an obstacle gives an equation for vertical velocity $w$

\begin{equation}
\frac{\partial^2 w}{\partial x^2} + \frac{\partial^2 w}{\partial z^2} + (\frac{N^2}{U^2} - \frac{1}{U(z)}\frac{dU(z)}{dz})w = 0
\end{equation}

This can be tackled by taking the Fourier transform for the case of constant $U$ and $N$.

\begin{equation}
-k^2 \hat{w} + \frac{\partial^2 \hat{w}}{\partial z^2} + \frac{N^2}{U^2}\hat{w} = 0
\end{equation}

where $\hat{w} = \hat{w}(k, z)$ and for each $k$ the vertical wavenumber is given by

\begin{equation}
m^2 = \frac{k^2(N^2 - \omega^2)}{\omega^2} = \frac{N^2}{U^2} - k^2
\end{equation}

For each wavenumber it is possible to obtain a solution for the vertical velocity as a function of depth. 

\begin{equation}
\hat{w}(k, z) = w_0(k) e^{imz}
\end{equation}

where the bottom boundary condition, $w(x, 0) = U \frac{dh}{dx}$, is first transformed to $\hat{w}(k, 0) = iUkh_k$, and then used to find the form of $W_0(k) = iUkh_k$. If $m$ is imaginary ($\omega^2 > N^2$) then the solution will decay with height. The real space solution can be retrieved by taking the inverse Fourier transform

\begin{equation}
w(x, z) = \sum_{k} iUkh_k e^{imz}
\end{equation}

I'm going to start by trying this for the Witch of Agnesi profile, like Bell does

\begin{equation}
h(x) = \frac{h_0 a^2}{x^2 + a^2}
\end{equation}

```python
import gravity_waves as gw
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline
% qtconsole
```

```python
# Set up the problem.
zT = 4e3
x, dx = np.linspace(-8e4, 8e4, 1000, retstep=True) # Horizontal dimension of domain [m]
z, dz = np.linspace(0, zT, 100, retstep=True) # Vertical dimension of domain [m]
# t, dt = np.linspace(0, 12.*60.*60., 2, retstep=True) # Time [s]
h_0 = 500. # Vertical height of mountain [m]
a = 10000. # Horizontal extent of mountain [m]
h = gw.witch_of_agnesi(x, a, h_0)

# Find the Fourier coefficients.

h_k = np.fft.fft(h)
k = np.fft.fftfreq(x.size, dx)
plt.figure()
plt.loglog(np.abs(k), np.abs(h_k))
plt.xlabel('$k$')

# Set the flow conditions.

U = 0.20 + 0j # Flow speed [m s-1] (adding 0j is a hack to make sqrt work better)
f = 0.000123 # Buoyancy frequency [s-1]
om = k*U # Frequency [s-1]
N = 0.0025 # Buoyancy frequency [s-1]
m = np.sqrt(k**2*(N**2 - om**2)/om**2)
# m = np.sqrt(k**2*(N**2 - om**2)/(om**2 - f**2))

# Bad hack
m[np.isnan(m)] = 0.

xg, zg = np.meshgrid(x, z)

idxs = k >= 0

u = np.zeros_like(xg)[:, idxs]
v = np.zeros_like(xg)
w = np.zeros_like(xg)[:, idxs]
p = np.zeros_like(xg)
w2 = np.zeros_like(xg)[:, idxs]
w3 = np.zeros_like(xg)[:, idxs]

# A = -k*(N**2 - om**2)/(m*om**2)



for i in xrange(len(z)):
    w[i, :] = np.real(np.fft.ifft(1j*k[idxs]*U*h_k[idxs]*np.exp(1j*m[idxs]*z[i])))
    u[i, :] = np.real(np.fft.ifft(-1j*m[idxs]*U*h_k[idxs]*np.exp(1j*m[idxs]*z[i])))
    
    wk = 1j*k[idxs]*U*h_k[idxs]*(np.exp(1j*m[idxs]*zi[i]) + 2j*np.sin(m[idxs]*zi[i])/(1 - np.exp(2j*m[idxs]*zT)))
    wk[np.isnan(wk)] = 0 + 0j
    w2[i, :] = np.real(np.fft.ifft(wk))
    
    wk = -1j*k[idxs]*h_k[idxs]*np.sin(m[idxs]*(zi[i] - zT))/np.sin(m[idxs])
    wk[np.isnan(wk)] = 0 + 0j
    w3[i, :] = np.real(np.fft.ifft(wk))
    
plt.figure()
plt.pcolormesh(xg[:, idxs]/1000.+40, zg[:, idxs], 100.*(u+U.real), cmap=plt.get_cmap('jet'))
# plt.fill(x/1000., h, 'k')
plt.xlim((x.min()/1000., x.max()/1000.))
plt.ylim((z.min(), z.max()))
plt.grid()
plt.xlabel('x (km)')
plt.ylabel('z (m)')
cbar = plt.colorbar(orientation='horizontal', extend='both')
cbar.set_label('u (cm s$^{-1}$)')
# plt.clim(-np.max(np.abs(u))*100., np.max(np.abs(u))*100.)

plt.figure()
plt.pcolormesh(xg[:, idxs]/1000.+40, zg[:, idxs], 100.*w, cmap=plt.get_cmap('bwr'))
# plt.fill(x/1000., h, 'k')
plt.xlim((x.min()/1000., x.max()/1000.))
plt.ylim((z.min(), z.max()))
plt.grid()
plt.xlabel('x (km)')
plt.ylabel('z (m)')
cbar = plt.colorbar(orientation='horizontal', extend='both')
cbar.set_label('w (cm s$^{-1}$)')
plt.clim(-np.max(np.abs(w))*100., np.max(np.abs(w))*100.)

plt.figure()
plt.pcolormesh(xg[:, idxs]/1000.+40, zg[:, idxs], 100.*w2, cmap=plt.get_cmap('bwr'))
# plt.fill(x/1000., h, 'k')
plt.xlim((x.min()/1000., x.max()/1000.))
plt.ylim((z.min(), z.max()))
plt.grid()
plt.xlabel('x (km)')
plt.ylabel('z (m)')
cbar = plt.colorbar(orientation='horizontal', extend='both')
cbar.set_label('w (cm s$^{-1}$)')
plt.clim(-np.max(np.abs(w2))*100., np.max(np.abs(w2))*100.)

plt.figure()
plt.pcolormesh(xg[:, idxs]/1000.+40, zg[:, idxs], 100.*w3, cmap=plt.get_cmap('bwr'))
# plt.fill(x/1000., h, 'k')
plt.xlim((x.min()/1000., x.max()/1000.))
plt.ylim((z.min(), z.max()))
plt.grid()
plt.xlabel('x (km)')
plt.ylabel('z (m)')
cbar = plt.colorbar(orientation='horizontal', extend='both')
cbar.set_label('w (cm s$^{-1}$)')
plt.clim(-np.max(np.abs(w3))*100., np.max(np.abs(w3))*100.)
```
