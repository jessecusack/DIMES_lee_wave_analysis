I am following the methodology presented by Vallis in the new chapters of his GFD book. It is as follows:

Consider a topographic profile $h(x)$ over a region of length $L$ which can be represented as a Fourier series

\begin{equation}
h(x) = \sum_k h_k e^{ikx}
\end{equation}

where $h_k$ are the fourier coefficients found by taking the fourier transform of $h(x)$. Presumably the horizontal wavenumbers are given by $k_n = \frac{2 \pi n}{L}$.

For each $k_n$ there is a vertical wavenumber given by

\begin{equation}
m^2 = \frac{k^2(N^2 - \omega^2)}{\omega^2}
\end{equation}

For each wavenumber it is possible to obtain a solution for the pressure $\phi_k(z)$ as a function of depth. This will be of the form

\begin{equation}
\phi_k(z) = \phi_0 e^{imz}
\end{equation}

where presumably if $m$ is complex then the solution will decay with height. This will happen if $\omega^2 > N^2$.

The real space solution should then be given as 

\begin{equation}
\phi(z) = \sum_{k, m} \phi_k e^{i(kx + mz)}
\end{equation}

where the amplitudes $\phi_k$ need to be found.

I'm going to start by trying this for the Witch of Agnesi profile, like Bell does

\begin{equation}
h(x) = \frac{h_0 a^2}{x^2 + a^2}
\end{equation}

```python
import gravity_waves as gw
import matplotlib.pyplot as plt
import numpy as np

% matplotlib inline
% qtconsole
```

```python
x, dx = np.linspace(-8e4, 8e4, 1000, retstep=True) # Horizontal dimension of domain [m]
z, dz = np.linspace(0, 5e3, 100, retstep=True) # Vertical dimension of domain [m]
t, dt = np.linspace(0, 12.*60.*60., 2, retstep=True) # Time [s]
h_0 = 1500. # Vertical height of mountain [m]
a = 300. # Horizontal extent of mountain [m]
L = x.max() - x.min()
h = gw.witch_of_agnesi(x, a, h_0)

plt.plot(x, h)
plt.xlim((x.min(), x.max()))
plt.ylim((z.min(), z.max()))
plt.grid()
plt.xlabel('x')
plt.ylabel('z')
```

```python
# Find the Fourier coefficients.

h_k = np.fft.fft(h)
k = np.fft.fftfreq(x.size, dx)
plt.loglog(np.abs(k), np.abs(h_k))
# plt.xlim((0, 0.0003))
```

```python
U = 0.5 + 0j # Flow speed [m s-1]
f = 0.000123 # Buoyancy frequency [s-1]
om = k*U # Frequency [s-1]
N = 0.0025 # Buoyancy frequency [s-1]
# m = np.sqrt(k**2*(N**2 - om**2)/om**2)
m = np.sqrt(k**2*(N**2 - om**2)/(om**2 - f**2))

xg, zg = np.meshgrid(x, z)

u = np.zeros_like(xg)
v = np.zeros_like(xg)
w = np.zeros_like(xg)
p = np.zeros_like(xg)

for hn, kn, mn, omn in zip(h_k, k, m, om):
    u += -np.real(1j*mn*U*hn*np.exp(1j*(kn*xg + mn*zg)))
    v += -np.real(1j*f*mn/kn*hn*np.exp(1j*(kn*xg + mn*zg)))
    w += np.real(1j*U*kn*hn*np.exp(1j*(kn*xg + mn*zg)))
    p += np.real(1j*mn*(U**2*kn**2 - f**2)/k**2*hn*np.exp(1j*(kn*xg + mn*zg)))

plt.figure()
plt.contour(xg, zg, u)
    
# plt.figure()
# plt.contour(xg, zg, v)
    
plt.figure()
plt.contour(xg, zg, w)

plt.figure()
plt.contour(xg, zg, p)
```
